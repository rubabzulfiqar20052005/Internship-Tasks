```python
import os
import csv
```

```python
folder_path = "student_records"

if not os.path.exists(folder_path):
    os.makedirs(folder_path)

print("Folder ready!")
```

```python
# Valid file
with open(os.path.join(folder_path, "classA.csv"), "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["name", "marks"])
    writer.writerow(["Ali", "78"])
    writer.writerow(["Sara", "45"])
    writer.writerow(["Ahmed", "88"])

# Corrupted file (invalid marks)
with open(os.path.join(folder_path, "classB.csv"), "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["name", "marks"])
    writer.writerow(["Zain", "abc"])   # Invalid marks
    writer.writerow(["Hina", "65"])

print("Sample files created!")
```

```python
class InvalidMarksError(Exception):
    pass
```

```python
clean_results = []

try:
    files = os.listdir(folder_path)

    for filename in files:
        file_path = os.path.join(folder_path, filename)

        try:
            with open(file_path, "r") as file:
                reader = csv.DictReader(file)

                for row in reader:
                    try:
                        name = row["name"]
                        marks = row["marks"]

                        if not marks.isdigit():
                            raise InvalidMarksError(f"Invalid marks for {name}")

                        marks = int(marks)

                        status = "Fail" if marks < 50 else "Pass"

                        clean_results.append({
                            "name": name,
                            "marks": marks,
                            "status": status
                        })

                    except InvalidMarksError as e:
                        print("Custom Error:", e)

                    except ValueError:
                        print("Value error in file:", filename)

                    except TypeError:
                        print("Type error in file:", filename)

        except FileNotFoundError:
            print("File not found:", filename)

except FileNotFoundError:
    print("Folder not found!")

finally:
    print("Processing Completed!")
```

```python
output_file = "clean_results.csv"

with open(output_file, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["name", "marks", "status"])

    for student in clean_results:
        writer.writerow([student["name"], student["marks"], student["status"]])

print("Clean results file created!")
```
