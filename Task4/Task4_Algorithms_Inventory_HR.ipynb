PART 1 – E-Commerce Inventory Analyzer

```python
# Imports
from typing import List, Dict
from collections import defaultdict
```

```python
products = [
    {"name": "Laptop", "category": "Electronics", "price": 1500},
    {"name": "Smartphone", "category": "Electronics", "price": 900},
    {"name": "Headphones", "category": "Electronics", "price": None},
    {"name": "Shoes", "category": "Fashion", "price": 120},
    {"name": "Shoes", "category": "Fashion", "price": 120},  # duplicate
    {"name": "T-Shirt", "category": "Fashion", "price": 35},
    {"name": "Coffee Maker", "category": "Home Appliances", "price": 80},
    {"name": "Blender", "category": "Home Appliances", "price": None},
    {"name": "Laptop", "category": "Electronics", "price": 1500},  # duplicate
    {"name": "Smartwatch", "category": "Electronics", "price": 200}
]

print(" Original Products:")
for p in products:
    print(p)
```

```python
# Remove duplicates & None prices using set of tuples
unique_products = { (p["name"], p["category"], p["price"]) for p in products if p["price"] is not None }

# Convert back to list of dicts
clean_products = [{"name": n, "category": c, "price": p} for (n,c,p) in unique_products]

print("\n Cleaned Products:")
for p in clean_products:
    print(p)
```

```python
category_dict = defaultdict(list)
[ category_dict[p["category"]].append(p) for p in clean_products ]
category_dict = dict(category_dict)

print("\n Products Grouped by Category:")
for cat, items in category_dict.items():
    print(f"{cat}:")
    for item in items:
        print(f"  {item}")
```

```python
top5_per_category = {
    cat: sorted(items, key=lambda x: x["price"], reverse=True)[:5]
    for cat, items in category_dict.items()
}

print("\n Top 5 Most Expensive Items per Category:")
for cat, items in top5_per_category.items():
    print(f"{cat}:")
    for item in items:
        print(f"  {item}")
```

```python
summary = {
    cat: {
        "total_items": len(items),
        "max_price": max(items, key=lambda x: x["price"])["price"]
    }
    for cat, items in category_dict.items()
}

print("\n Summary:")
for cat, s in summary.items():
    print(f"{cat}: Total Items={s['total_items']}, Max Price={s['max_price']}")
```

PART 2 – HR Employee Ranker

```python
employees = [
    {"name": "Ali", "score": 90, "experience": 5},
    {"name": "Sara", "score": 95, "experience": 3},
    {"name": "Zain", "score": 90, "experience": 7},
    {"name": "Hina", "score": 85, "experience": 4},
    {"name": "Ahmed", "score": 90, "experience": 5}
]

print("\n Original Employee List:")
for e in employees:
    print(e)
```

```python
ranked_employees = sorted(
    employees,
    key=lambda x: (-x["score"], -x["experience"], x["name"])
)

print("\n Ranked Employees:")
for e in ranked_employees:
    print(e)
```

```python
def binary_search_employee(sorted_list: List[Dict], name: str) -> Dict:
    low, high = 0, len(sorted_list) - 1
    while low <= high:
        mid = (low + high) // 2
        if sorted_list[mid]["name"] == name:
            return sorted_list[mid]
        elif sorted_list[mid]["name"] < name:
            low = mid + 1
        else:
            high = mid - 1
    return None
```

```python
search_name = "Zain"
found_employee = binary_search_employee(ranked_employees, search_name)

print(f"\n Search Result for {search_name}:")
print(found_employee)
```
