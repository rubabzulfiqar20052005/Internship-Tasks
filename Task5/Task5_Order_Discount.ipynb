```python
from functools import reduce
```

```python
orders = [
    {"customer": "Ali", "category": "Electronics", "amount": 1500},
    {"customer": "Sara", "category": "Fashion", "amount": 200},
    {"customer": "Hina", "category": "Electronics", "amount": 900},
    {"customer": "Ahmed", "category": "Fashion", "amount": 50},
    {"customer": "Zain", "category": "Home Appliances", "amount": 300},
    {"customer": "Lara", "category": "Electronics", "amount": 400},
    {"customer": "Nina", "category": "Fashion", "amount": 150}
]

print(" Raw Orders:")
for o in orders:
    print(o)
```

```python
discount_rules = {
    "Electronics": 0.10,  # 10% discount
    "Fashion": 0.20,      # 20% discount
    "Home Appliances": 0.15
}

min_threshold = 100  # Orders below this after discount will be filtered out
```

```python
discounted_orders = list(map(
    lambda o: {**o, "final_amount": o["amount"] * (1 - discount_rules.get(o["category"], 0))},
    orders
))

print("\n Orders After Discount:")
for o in discounted_orders:
    print(o)
```

```python
filtered_orders = list(filter(
    lambda o: o["final_amount"] >= min_threshold,
    discounted_orders
))

print("\n Orders After Filtering (Final Amount >= 100):")
for o in filtered_orders:
    print(o)
```

```python
sorted_orders = sorted(filtered_orders, key=lambda o: o["final_amount"], reverse=True)

print("\n Orders Sorted by Final Amount (High → Low):")
for o in sorted_orders:
    print(o)
```

```python
total_revenue = reduce(
    lambda acc, o: acc + o["final_amount"],
    sorted_orders,
    0
)

print(f"\n Total Revenue After Discounts: Rs {total_revenue}")
```

```python
summary = {
    o["customer"]: {
        "category": o["category"],
        "original_amount": o["amount"],
        "final_amount": o["final_amount"]
    }
    for o in sorted_orders
}

print("\n Processed Summary:")
for customer, info in summary.items():
    print(f"{customer}: {info}")
```
