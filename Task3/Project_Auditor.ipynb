```python
scanner_code = """
import os
import math
from datetime import datetime

PENALTY_PER_DAY = 20

def scan_directory(folder_path):

    audit_data = []
    penalties = []

    if not os.path.exists(folder_path):
        raise FileNotFoundError("Folder does not exist!")

    current_time = datetime.now()

    for filename in os.listdir(folder_path):

        file_path = os.path.join(folder_path, filename)

        if os.path.isfile(file_path):

            stats = os.stat(file_path)

            size_kb = math.ceil(stats.st_size / 1024)

            modified_time = datetime.fromtimestamp(stats.st_mtime)

            age_days = (current_time - modified_time).days

            file_info = {
                "name": filename,
                "size_kb": size_kb,
                "type": filename.split('.')[-1],
                "last_modified": modified_time,
                "age_days": age_days,
                "flagged": age_days > 30
            }

            audit_data.append(file_info)

            if age_days > 30:
                overdue = age_days - 30
                fine = overdue * PENALTY_PER_DAY

                penalties.append({
                    "name": filename,
                    "overdue_days": overdue,
                    "fine": fine
                })

    return audit_data, penalties
"""

with open("auditor/scanner.py", "w") as f:
    f.write(scanner_code)

print("Custom module created!")
```

```python
timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
timestamp
```

```python
audit_filename = f"audit_report_{timestamp}.txt"

with open(audit_filename, "w") as file:

    file.write("PROJECT DIRECTORY AUDIT REPORT\n")
    file.write("=" * 60 + "\n\n")

    total_size = 0
    flagged_count = 0

    for item in audit_data:

        total_size += item["size_kb"]
        if item["flagged"]:
            flagged_count += 1

        file.write(f"File: {item['name']}\n")
        file.write(f"Type: {item['type']}\n")
        file.write(f"Size: {item['size_kb']} KB\n")
        file.write(f"Age: {item['age_days']} days\n")
        file.write(f"Flagged: {item['flagged']}\n")
        file.write("-" * 40 + "\n")

    file.write("\nSUMMARY\n")
    file.write("=" * 40 + "\n")
    file.write(f"Total Files: {len(audit_data)}\n")
    file.write(f"Flagged Files: {flagged_count}\n")
    file.write(f"Total Size (KB): {total_size}\n")

print("Audit report generated!")
```

```python
penalty_filename = f"penalties_report_{timestamp}.txt"

with open(penalty_filename, "w") as file:

    file.write("PENALTIES REPORT\n")
    file.write("=" * 50 + "\n\n")

    total_fine = 0

    for p in penalties:
        total_fine += p["fine"]

        file.write(f"File: {p['name']}\n")
        file.write(f"Overdue Days: {p['overdue_days']}\n")
        file.write(f"Fine: Rs {p['fine']}\n")
        file.write("-" * 40 + "\n")

    file.write(f"\nTOTAL FINE: Rs {total_fine}")

print("Penalty report generated!")
```
