```python
from datetime import datetime, timedelta
```

```python
class Book:

    def __init__(self, title, author):

        self.title = title
        self.author = author
        self.is_borrowed = False
        self.borrow_date = None
        self.due_date = None


    def __str__(self):

        status = "Borrowed" if self.is_borrowed else "Available"

        return f"Book: {self.title} | Author: {self.author} | Status: {status}"
```

```python
class Member:

    def __init__(self, name):
        self.name = name
        self.borrowed_books = []

    def borrow_limit(self):
        return 2  # Default borrow limit for normal members

    def borrow_book(self, book):
        if book.is_borrowed:
            print("Book already borrowed")
            return

        if len(self.borrowed_books) >= self.borrow_limit():
            print("Borrow limit reached")
            return

        book.is_borrowed = True
        book.borrow_date = datetime.now()
        book.due_date = datetime.now() + timedelta(days=14)

        self.borrowed_books.append(book)
        print(f"{self.name} borrowed {book.title}")
        print(f"Due date: {book.due_date.date()}")

    def return_book(self, book):
        if book not in self.borrowed_books:
            print("Book not borrowed by this member")
            return

        today = datetime.now()

        if today > book.due_date:
            days_late = (today - book.due_date).days
            fine = self.calculate_fine(days_late)
            print(f"Book returned late by {days_late} days")
            print(f"Fine: {fine}")

        book.is_borrowed = False
        book.borrow_date = None
        book.due_date = None

        self.borrowed_books.remove(book)
        print(f"{self.name} returned {book.title}")

    def calculate_fine(self, days):
        return days * 5  # Default fine for normal members

    def show_borrowed_books(self):
        if not self.borrowed_books:
            print("No books borrowed")
            return

        for book in self.borrowed_books:
            print(f"{book.title} | Due: {book.due_date.date()}")

    def __str__(self):
        return f"Member: {self.name} | Borrowed Books: {len(self.borrowed_books)}"
```

```python
class PremiumMember(Member):
    def borrow_limit(self):
        return 5
    def calculate_fine(self, days):
        return days * 2
```

```python
class EliteMember(PremiumMember):
    def borrow_limit(self):
        return 10
    def calculate_fine(self, days):
        return days * 1
```

```python
class Library:

    def __init__(self):

        self.books = []
        self.members = []


    def add_book(self, book):

        self.books.append(book)
        print(f"Book added: {book.title}")


    def register_member(self, member):

        self.members.append(member)
        print(f"Member registered: {member.name}")

    def show_borrowed_books(self):

       if not self.borrowed_books:
        print("No books borrowed")
        return

       for book in self.borrowed_books:

        print(
            f"{book.title} | Due: {book.due_date.date()}"
        )
    def show_books(self):

        for book in self.books:
            print(book)
    def search_book(self, keyword):

      found = False

      for book in self.books:

        if keyword.lower() in book.title.lower():

            print(book)
            found = True

      if not found:
        print("No book found")
```
