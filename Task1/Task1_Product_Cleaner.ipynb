**Objective:** Clean raw product data, remove duplicates, convert prices to float, group by category and print a neat summary.

```python
# Raw product data
products = [
    {"name": "Laptop", "price": "1000", "category": "Electronics"},
    {"name": "Shirt", "price": None, "category": "Clothing"},
    {"name": "Laptop", "price": "1000", "category": "Electronics"},
    {"name": "Headphones", "price": "150", "category": "Electronics"},
    {"name": "Jeans", "price": "50", "category": "Clothing"},
    {"name": "Coffee Mug", "price": "10.5", "category": "Home"},
    {"name": "Coffee Mug", "price": "10.5", "category": "Home"},
    {"name": "Notebook", "price": "abc", "category": "Stationery"}
]
```

```python
cleaned_products = []
seen = set()

for product in products:
    try:
        name = product['name']
        category = product['category']
        price = float(product['price'])  # Convert price to float
        identifier = (name, price, category)
        if identifier not in seen:
            seen.add(identifier)
            cleaned_products.append({
                'name': name,
                'price': price,
                'category': category
            })
    except (TypeError, ValueError):
        # Skip products with missing or invalid price
        continue
```

```python
from collections import defaultdict

category_dict = defaultdict(list)
for product in cleaned_products:
    category_dict[product['category']].append({
        'name': product['name'],
        'price': product['price']
    })
```

```python
for category, items in category_dict.items():
    print(f"\nCategory: {category}")
    for item in items:
        print(f"  {item['name']} - ${item['price']}")
    print(f"  Total Items: {len(items)}")
```
