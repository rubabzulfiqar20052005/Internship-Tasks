```python
from collections import defaultdict
import itertools
```

```python
# Simulated server logs
log_lines = [
    "INFO: Server started",
    "ERROR: Database connection failed",
    "WARNING: Disk space low",
    "ERROR: Timeout occurred",
    "INFO: Request processed",
    "ERROR: Database connection failed",
    "ERROR: Out of memory",
    "WARNING: High CPU usage",
    "ERROR: Timeout occurred",
    "INFO: Server running smoothly"
]

print("🖥 Sample Logs:")
for line in log_lines:
    print(line)
```

```python
def critical_error_stream(log_iterable):
    """
    Generator function to yield only ERROR lines from logs.
    """
    for line in log_iterable:
        if line.startswith("ERROR:"):
            yield line
```

```python
error_counter = defaultdict(int)

print("\n Streaming Critical Errors and Counting Live:")
for error in critical_error_stream(log_lines):
    error_type = error.replace("ERROR: ", "")
    error_counter[error_type] += 1
    print(f"{error} | Running Count: {error_counter[error_type]}")
```

```python
# error_counter already contains counts per error type
print("\n Summary of Critical Errors by Type:")
for err_type, count in error_counter.items():
    print(f"{err_type}: {count}")
```

```python
# Imagine multiple log files
log_file1 = [
    "ERROR: Disk failure",
    "INFO: Backup completed",
]

log_file2 = [
    "ERROR: Disk failure",
    "ERROR: Network unreachable",
]

all_logs = itertools.chain(log_file1, log_file2, log_lines)  # chain multiple sources

# Streaming with same generator
error_counter_chain = defaultdict(int)
print("\n Streaming Combined Logs:")
for error in critical_error_stream(all_logs):
    err_type = error.replace("ERROR: ", "")
    error_counter_chain[err_type] += 1
    print(f"{error} | Running Count: {error_counter_chain[err_type]}")
```
